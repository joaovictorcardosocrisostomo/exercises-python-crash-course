## Boas práticas de escrita de nomes para variáveis:
• Nomes de variáveis podem conter apenas letras, números e
underscores. Podem começar com uma letra ou um underscore, mas
não com um número. Por exemplo, podemos chamar uma variável de
message_1, mas não de 1_message.

• Espaços não são permitidos em nomes de variáveis, mas underscores
podem ser usados para separar palavras em nomes de variáveis. Por
exemplo, greeting_message funciona, mas greeting message causará
erros.

• Evite usar palavras reservadas e nomes de funções em Python como
nomes de variáveis, ou seja, não use palavras que Python reservou para
um propósito particular de programação, por exemplo, a palavra print.
(Veja a seção “Palavras reservadas e funções embutidas de Python”.)

• Nomes de variáveis devem ser concisos, porém descritivos. Por
exemplo, name é melhor que n, student_name é melhor que s_n e
name_length é melhor que length_of_persons_name.

• Tome cuidado ao usar a letra minúscula l e a letra maiúscula O, pois
elas podem ser confundidas com os números 1 e 0.

## Strings
Uma string é simplesmente uma série de caracteres. Tudo que estiver
entre aspas é considerada uma string em Python, e você pode usar aspas
simples ou duplas em torno de suas strings, assim:
"This is a string."
'This is also a string.'

Essa flexibilidade permite usar aspas e apóstrofos em suas strings:
'I told my friend, "Python is my favorite language!"'
"The language 'Python' is named after Monty Python, not the snake."
"One of Python's strengths is its diverse and supportive community."
Vamos explorar algumas das maneiras de usar strings.

### Convertendo letras minúsculas em maiúsculas

```python
""" a variável name recebe a string """
name = "alessandra silva"
""" o método title() aparece na instrução print() depois de a variável ser definida """
print(name.title())
```

Um método é uma ação que python pode executar em um dado. O ponto (.) indica que o método deve ser usado na variável name. Todo método é seguido de parênteses, pois costumam precisar de informações adicionais. Como o método title não precisa de informações, não há nada entre os parênteses.
Esse método exibe a string com uma letra maiúscula no início de cada palavra. Por exemplo:

```python
name = input("Escreva algo com letras maiúsculas, minúsculas ou misturadas: ")
""" mesmo que eu digite só letras maiúsculas ou só letras minúsculas, esse método trata o dado 
para que apenas a primeira letra fique maiúscula e todo o resto seja composto por letras minúsculas """
print(name.title())
```

Outros métodos que existem para tratamento de maiúsculas e minúsculas são upper() e lower(). Veja:

```python
word = input("Escreva algo com letras maiúsculas, minúsculas ou misturadas: ")
print(word.upper())
print(word.lower())
```

### Combinando ou concatenando strings
Veja:

```python
first_name = "alessandra"
last_name = "silva"
full_name = first_name + " " + last_name
""" ou até """
name = first_name.title() + " " + last_name.title()
print(full_name)
print(name)
```

Outro caso seria:

```python
first_name = "alessandra"
last_name = "silva"
full_name = first_name + " " + last_name
print("Hello, " + full_name.title() + "!")

""" ou até """
message = "Hello, " + full_name.title() + "!"
print(message)
```

Para acrescentar uma tabulação a uma string, use a combinação decaracteres \t. Veja:

```python
print("python")
print("\tpython")
```

Ou uma quebra de linha com \n:

```python
print("python\n")
print("Languages: \nPython\nC\nJava")
```

Esses comandos também são aninháveis. Observe:

```python
print("Languages: \n\tPython\n\tC\n\tJava\n")
```

### Removendo espaços em branco
Espaços em branco extras podem ser confusos em seus programas. Para os
programadores, 'python' e 'python ' parecem praticamente iguais.
Contudo, para um programa, são duas strings diferentes. Python identifica
o espaço extra em 'python ' e o considera significativo, a menos que você
informe o contrário.

No caso, por exemplo, de um usuário preenchendo seu nome com excesso de espaços em branco à esquerda ou à direita num site, há como tratar com rstrip(), lstrip() ou strip() (para ambos).
Um exemplo:

```python
name = "joao "
user = name.rstrip()
print(user)
```

Para fazer a persistência dessa modificação, a variável recebe o resultado do próprio método:

```python
name = " joao "
name = name.rstrip()
print(name)
name = name.lstrip()
print(name)

# para remover dos dois lados ao mesmo tempo
name = " joao "
name = name.strip()
print(name)
```

## Números
Existem tipos diferentes de números que vão ser tratados de maneira diferente em python. Veja:

### Inteiros
Vamos começar o estudo dos números pelos inteiros. Você pode somar (+), subtrair (-), multiplicar (*) e dividir (/) inteiros em Python.

```python
num_1 = int(input("digite um número: "))
num_2 = int(input("digite outro número: "))
soma = num_1 + num_2
sub = num_1 - num_2
multi = num_1 * num_2
div = num_1 / num_2
exp = num_1 ** num_2

print(f'Somados: {soma}\n')
print(f'Subtraídos: {sub}\n')
print(f'Multiplicados: {multi}\n')
print(f'Divididos: {div}\n')
print(f'O segundo como exp do primeiro: {exp}\n')
```

Python reconhece a ordem das operações matemáticas. Espaços em branco não afetam em nada o código, apenas ajudam o dev a identificar a ordem de prioridade. Veja: 
>>> 2 + 3*4
14
>>> (2 + 3) * 4
20

### Float (Ponto Flutuante)
Python chama qualquer número com um ponto decimal de número de
ponto flutuante (float). Esse termo é usado na maioria das linguagens de
programação e refere-se ao fato de um ponto decimal poder aparecer em
qualquer posição em um número. Toda linguagem de programação deve
ser cuidadosamente projetada para lidar de modo adequado com números
decimais de modo que eles se comportem de forma apropriada, não
importa em que lugar o ponto decimal apareça.
Na maioria das ocasiões, podemos usar decimais sem nos preocupar com
o modo como eles se comportam. Basta fornecer os números que você quer
usar, e Python provavelmente fará o que você espera:
>>> 0.1 + 0.1
0.2
>>> 0.2 + 0.2
0.4
>>> 2 * 0.1
0.2
>>> 2 * 0.2
0.4

No entanto, tome cuidado, pois, às vezes, você poderá obter um número
arbitrário de casas decimais em sua resposta:
>>> 0.2 + 0.1
0.30000000000000004
>>> 3 * 0.1
0.30000000000000004

Isso acontece em todas as linguagens e não é motivo para muita
preocupação. É coisa de como a memória se comporta.

## Evitando erros de tipo com a função str()
Você poderia escrever um código como este:
birthday.py
age = 23
message = "Happy " + age + "rd Birthday!"
print(message)

Você esperaria que esse código exibisse a seguinte saudação simples de
feliz aniversário: Happy 23rd birthday!. 

Contudo, se executar esse código,
verá que ele gera um erro:
Traceback (most recent call last):
File "birthday.py", line 2, in <module>
message = "Happy " + age + "rd Birthday!"
u TypeError: Can't convert 'int' object to str implicitly

É um erro de tipo. Significa que Python não é capaz de reconhecer o tipo
de informação que você está usando.
O interpretador poderia imprimir tanto o número 23 quando os caractéres 2 e 3. Assim, para tratar o erro, basta especificar como você quer que a variável seja apresentada (como string). Veja:

```python
age = 23
message = "Happy " + str(age) + "rd Birthday!"
print(message)
```

```python
# Aplicando essa correção no código que eu tratei de outra maneira acima
num_1 = int(input("digite um número: "))
num_2 = int(input("digite outro número: "))
soma = num_1 + num_2
sub = num_1 - num_2
multi = num_1 * num_2
div = num_1 / num_2
exp = num_1 ** num_2

print('Somados: ' + str(soma) + '\n')
print('Subtraídos: ' + str(sub) + '\n')
print('Multiplicados: ' + str(multi) + '\n')
print('Divididos: ' + str(div) + '\n')
print('O segundo como exp do primeiro: ' + str(exp) + '\n')
```

Com isso, Python sabe que você quer converter o valor numérico em
uma string e exibir os caracteres como parte da mensagem.
